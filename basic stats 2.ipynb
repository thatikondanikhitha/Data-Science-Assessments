#### Estimation And Confidence Intervals

```python
import numpy as np
import scipy.stats as stats
```

```python
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 
                 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
# finding sample size,mean,std
n=len(data)
mean = np.mean(data)
sample_std = np.std(data, ddof=1) 
print("Sample size (n):", n)
print("Sample mean:", round(mean, 3))
print("Sample standard deviation (s):", round(sample_std, 3))
```

##### a. Build 99% Confidence Interval Using Sample Standard Deviation

```python
confidence = 0.99
df = n - 1  # degrees of freedom
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df)  # t critical value

# Standard error
se_t = sample_std / np.sqrt(n)

# Margin of error
margin_t = t_critical * se_t

# Confidence interval
ci_t_lower = mean - margin_t
ci_t_upper = mean + margin_t

print(f"\n(a) 99% Confidence Interval using t-distribution:")
print(f"t-critical value: {t_critical:.3f}")
print(f"Standard Error: {se_t:.3f}")
print(f"Confidence Interval: ({ci_t_lower:.3f}, {ci_t_upper:.3f})")
```
