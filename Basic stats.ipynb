#### ASSISSMENT ON BASICS

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("sales_data_with_discounts.csv") # Lading the data set
data.head()
```

```python
data.info() # describing the numeric columns in dataset
```

```python
data = data.select_dtypes(include=['int64', 'float64'])
print("Numerical_columns: ",data.columns.tolist())
```

```python
mean = np.mean(data)
print(mean)
```

```python
median = np.median(data)

std_dev = np.std(data)

print(mean, median, std_dev)
```

```python
#calculating mean,median,std of avg_price
mean_Avg_price = data["Avg Price"].mean()
median_Avg_price = data["Avg Price"].median()
standarddeviation_Avg_price = data["Avg Price"].std()
print("Mean:",mean_Avg_price,
      "Median:",median_Avg_price,
      "STD:",standarddeviation_Avg_price)
```

```python
#calculating mean median and std of total_sales
mean_Total_Sales_Value = data["Total Sales Value"].mean()
median_Total_Sales_Value = data["Total Sales Value"].median()
standarddeviation_Total_Sales_Value = data["Total Sales Value"].std()
print("Mean:",mean_Total_Sales_Value,
      "Median:",median_Total_Sales_Value,
      "STD:",standarddeviation_Total_Sales_Value)
```

```python
# Calculating the  mean, median, mode, and standard deviation
stats_values = pd.DataFrame({
    'Mean': data.mean(),
    'Median': data.median(),
    'Mode': [data[col].mode()[0] if not data[col].mode().empty else None for col in data.columns],
    'Standard Deviation':data.std()
})

print(stats_values)
```

```python
import math
```

```python
num_cols = data.columns
num_plots = len(num_cols)

# Layout: max 3 plots per row
cols_per_row = 2
rows = math.ceil(num_plots / cols_per_row)

plt.figure(figsize=(6 * cols_per_row, 5 * rows))

for i, col in enumerate(num_cols):
    plt.subplot(rows, cols_per_row, i + 1)
    sns.boxplot(y=data[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

#### Bar chart

```python
cat_cols = data.columns
num_plots = len(cat_cols)

# Layout: max 3 plots per row
cols_per_row = 3
rows = math.ceil(num_plots / cols_per_row)

plt.figure(figsize=(6 * cols_per_row, 5 * rows))

for i, col in enumerate(cat_cols):
    plt.subplot(rows, cols_per_row, i + 1)
    sns.countplot(x=data[col], palette='pastel', order=data[col].value_counts().index)
    plt.title(f'Bar Chart of {col}')
    plt.xticks(rotation=45)  # Rotate labels if needed

plt.tight_layout()
plt.show()
```

#### Standardization of Numerical Variables



```python
standardized_df = (data - data.mean()) / data.std()


print("Original Data (first 5 rows):")
print(data.head())

print("\nStandardized Data (first 5 rows):")
print(standardized_df.head())
```

#### Compare distributions before and after standardization

```python
num_cols = data.columns
plt.figure(figsize=(15, 5 * len(num_cols)))

for i, col in enumerate(num_cols, 1):
    plt.subplot(len(num_cols), 2, 2*i - 1)
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f'Original: {col}')
    
    plt.subplot(len(num_cols), 2, 2*i)
    sns.histplot(standardized_df[col], kde=True, color='salmon')
    plt.title(f'Standardized: {col}')

plt.tight_layout()
plt.show()
```
