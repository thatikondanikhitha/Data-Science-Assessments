#### 2. Dimensionality Reduction with PCA

#### Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
#apply PCA
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(scaled_data)
explained_var = pca.explained_variance_ratio_
cum_var = np.cumsum(explained_var)
plt.figure(figsize=(10,6))
plt.bar(range(1, len(explained_var)+1), explained_var, alpha=0.5, align='center', label='individual')
plt.step(range(1, len(cum_var)+1), cum_var, where='mid', label='cumulative')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Components')
plt.title('Scree Plot')
plt.legend(loc='best')
plt.show()
```

```python
n_components = np.where(cum_var >= 0.9)[0][0] + 1  # Minimum components for ~90% variance
print(f'Using {n_components} components')
pca_final = PCA(n_components=n_components)
X_pca_final = pca_final.fit_transform(scaled_data)
```

```python
#Transforming the data into PCA
X_pca = pca.transform(scaled_data)
print("Original shape:",scaled_data.shape)
print("Transformed shape:", X_pca.shape)
```

```python
#data frame with PCA
pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(pca.n_components_)])
print(pca_df.head())
```

#### 3.Clustering with Original Data

##### Kmeans on original Data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
kmeans = KMeans(n_clusters=4, random_state=42)
labels_orig = kmeans.fit_predict(scaled_data)
data['cluster_orig'] = labels_orig
```

##### Visualization

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=scaled_data[:,0], y=scaled_data[:,1], hue=labels_orig, palette='Set1')
plt.title('KMeans Clusters (Original Data)')
plt.xlabel(num_cols[0])
plt.ylabel(num_cols[1])
plt.legend(title='Cluster')
plt.show()
```

##### comparasion

```python
print('Silhouette Score (original):', silhouette_score(scaled_data, labels_orig))
print('Davies-Bouldin Index (original):', davies_bouldin_score(scaled_data, labels_orig))
```

#### 4.Clustering with PCA Data
##### KMeans on PCA Data

```python
kmeans_pca = KMeans(n_clusters=4, random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca_final)
data['cluster_pca'] = labels_pca
```

##### visualization

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x=X_pca_final[:,0], y=X_pca_final[:,1], hue=labels_pca, palette='Set2')
plt.title('KMeans Clusters (PCA Data)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Cluster')
plt.show()
```

##### comparasion

```python
print('Silhouette Score (PCA):', silhouette_score(X_pca_final, labels_pca))
print('Davies-Bouldin Index (PCA):', davies_bouldin_score(X_pca_final, labels_pca))
```
