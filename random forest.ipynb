#### Data Processing

```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
# Convert to numeric where possible
for col in data.columns:
    try:
        data[col] = pd.to_numeric(data[col])
    except:
        pass
```

```python
print("Missing values in each column:")
print(data.isnull().sum())

# If missing values exist, fill with mean of each column
if data.isnull().sum().any():
    data = data.fillna(data.mean())
    print("\nMissing values were imputed using column mean.")
else:
    print("\nNo missing values found in dataset.")
```

```python
# Check for categorical variables (none expected)

cat_cols = data.select_dtypes(include=['object']).columns
print("\nCategorical columns:", list(cat_cols))

if len(cat_cols) > 0:
    data = pd.get_dummies(data, columns=cat_cols, drop_first=True)
    print("\nApplied One-Hot Encoding to categorical variables.")
else:
    print("\nNo categorical columns found. Skipping encoding step.")
```

```python
#Feature Scaling (Standardization)

# Separate features and target variable
X = data.iloc[:, :-1]  # all columns except last one
y = data.iloc[:, -1]   # last column (target)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nFeature scaling completed using StandardScaler.")
```

```python
#Handle Imbalanced Data using SMOTE

from collections import Counter
print("\nBefore SMOTE, class distribution:")
print(Counter(y))

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

print("\nAfter SMOTE, class distribution:")
print(Counter(y_resampled))
```

```python
print("\nFinal feature matrix shape:", X_resampled.shape)
print("Final target vector shape:", y_resampled.shape)
```

#### Random forest Model Implementation

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Train-Test split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

print("\nTraining set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

```python
# Build Random Forest Model
rf = RandomForestClassifier(
    n_estimators=100,      # number of trees
    max_depth=None,        # let trees expand fully
    random_state=42
)
rf.fit(X_train, y_train)
```

```python
# Model Evalutation
y_pred = rf.predict(X_test)

# Compute metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("\n--- Random Forest Performance ---")
print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1 Score : {f1:.4f}")
```

```python
# Detailed classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Confusion matrix heatmap
plt.figure(figsize=(7,5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title("Random Forest  Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
#Feature Importance (optional but insightful)

importances = rf.feature_importances_
feature_names = X.columns
feat_imp = pd.Series(importances, index=feature_names).sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=feat_imp, y=feat_imp.index, palette='viridis')
plt.title("Feature Importance in Random Forest")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```

#### Bagging and Boosting Methods

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    BaggingClassifier,
    AdaBoostClassifier,
    GradientBoostingClassifier
)
```

```python
#Random Forest (Bagging-based ensemble)

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
```

```python
# Bagging Classifier

bag = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=50,
    random_state=42
)
bag.fit(X_train, y_train)
bag_pred = bag.predict(X_test)
bag_acc = accuracy_score(y_test, bag_pred)
```

```python
#AdaBoost Classifier (Boosting)

ada = AdaBoostClassifier(
    n_estimators=100,
    random_state=42
)
ada.fit(X_train, y_train)
ada_pred = ada.predict(X_test)
ada_acc = accuracy_score(y_test, ada_pred)
```

```python
#Gradient Boosting Classifier (Boosting)

gb = GradientBoostingClassifier(
    n_estimators=100,
    random_state=42
)
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)
gb_acc = accuracy_score(y_test, gb_pred)
```

```python
#Compare Results
# -----------------------------------------------------------
results = pd.DataFrame({
    "Model": ["Random Forest", "Bagging", "AdaBoost", "Gradient Boosting"],
    "Accuracy": [rf_acc, bag_acc, ada_acc, gb_acc]
})

print("\n--- Model Performance Comparison ---")
print(results)
```

```python
# Plot comparison
plt.figure(figsize=(7,5))
sns.barplot(x="Model", y="Accuracy", data=results, palette="viridis")
plt.title("Comparison of Ensemble Methods")
plt.ylim(0,1)
plt.show()
```

```python
#Detailed evaluation for the best model
best_model = results.loc[results['Accuracy'].idxmax(), 'Model']
print(f"\nBest Performing Model: {best_model}")

if best_model == "Random Forest":
    y_best = rf_pred
elif best_model == "Bagging":
    y_best = bag_pred
elif best_model == "AdaBoost":
    y_best = ada_pred
else:
    y_best = gb_pred

print("\nClassification Report for Best Model:\n")
print(classification_report(y_test, y_best))
sns.heatmap(confusion_matrix(y_test, y_best), annot=True, fmt='d', cmap='coolwarm')
plt.title(f"{best_model} - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
