#### 4: Word counter

```python
# Program: Word Counter

# Input text
input= "This text will be used to demonstrate the word counter."

# Split the text into words
words = input.split()

# Create an empty dictionary to store word counts
word_count = {}

# Looping through each word in the list
for word in words:
    # Convert to lowercase to make counting case-insensitive
    word = word.lower()
    # Increase the count for the word
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Display the word counts
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

#### 5: Check for Palindrome



```python
#program check for palindrome
import string
def is_palindrome(text):
    # Convert to lowercase to ignore case
    text = text.lower()
    
    # Remove spaces and punctuation
    text = ''.join(ch for ch in text if ch.isalnum())
    
    # Check if the text reads the same forward and backward
    return text == text[::-1]


# Example usage
input_text = "racecar"
print(is_palindrome(input_text))
print("Given input_text is a palindrome")
# Expected Output: True
```
