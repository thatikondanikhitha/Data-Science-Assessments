```python
# Import required libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Load dataset
df = pd.read_excel("Online retail.xlsx")

# The dataset contains one column with comma-separated items
print(df.head())

# Convert each row into a list of items
transactions = df.iloc[:, 0].apply(lambda x: x.split(','))

# Initialize Transaction Encoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm to find frequent itemsets
frequent_items = apriori(df_encoded, min_support=0.02, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_items, metric="lift", min_threshold=1.0)

# Sort and show top rules
rules = rules.sort_values(by=['lift', 'confidence'], ascending=False)
print(rules.head(10))
```
